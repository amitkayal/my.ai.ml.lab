```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
dataset =  pd.read_csv('data/startups.csv')
```

```python
dataset
```

```python
sns.set()
# Use Square root to decide bins
_ = plt.hist(dataset['Profit'], bins=int(np.sqrt(dataset.size))) 
_ = plt.xlabel('Profit')
_ = plt.ylabel('# of companies')
plt.show()
```

Binning bias: The same data may be interpreted differently depending on choice of bins

Solution: Bee swarm plot

```python
_ = sns.swarmplot(x='State', y='Profit', data=dataset)
plt.xlabel('State')
plt.ylabel('Profit')
plt.show()
```
